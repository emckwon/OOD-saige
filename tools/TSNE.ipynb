```python
import torch
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import os
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
model = TSNE(learning_rate=100)
```

```python
root = "/home/sr2/Hyeokjun/OOD-saige/results"
expdir = "res18_pretrained_severstal012_isomax"
ood_dataset = "classifier"
path = os.path.join(root, expdir, "valid", ood_dataset)
feature = torch.load(os.path.join(path, "logits.pt")).numpy()
labels = torch.load(os.path.join(path, "targets.pt")).numpy()
```

```python
#isomax aug
transformed = model.fit_transform(feature)
pc_y = np.c_[transformed,labels]
df = pd.DataFrame(pc_y,columns=['PC1','PC2','h'])
sns.scatterplot(data=df, x='PC1',y='PC2', hue='h')
```
