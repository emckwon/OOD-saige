```python
import torch
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
root = "/home/sr2/Hyeokjun/OOD-saige/results"
expdir = "res18_pretrained_cifar10_isomaxpush"
ood_dataset = "tinyimagenet"
path = os.path.join(root, expdir, "valid", ood_dataset)
logits = torch.load(os.path.join(path, "logits.pt")).numpy()
targets = torch.load(os.path.join(path, "targets.pt")).numpy()
```

```python
X_ = StandardScaler().fit_transform(logits)
pca = PCA(n_components=2)
pc = pca.fit_transform(X_)
```

```python
pc_y = np.c_[pc,targets]
```

```python
df = pd.DataFrame(pc_y,columns=['PC1','PC2','h'])
#sns.scatterplot(data=df,x='PC1',y='PC2',hue='diagnosis')
sns.scatterplot(data=df,x='PC1',y='PC2',hue='h')
```
